```python
import ai4water
from ai4water.datasets import MtropicsLaos
from ai4water.eda import EDA
```

```python
laos = MtropicsLaos()
```

# Precipitation

```python
pcp = laos.fetch_pcp()
```

```python
pcp.shape
```

```python
eda = EDA(pcp)
```

```python
eda.plot_missing(save=False, figsize=(3,5))
```

So around 5% of precipitation data is missing.

Find location of missing values in the data.

```python
eda.heatmap(save=False)
```

```python
eda.plot_data(save=False, figsize=(10, 6))
```

```python
eda.plot_histograms(save=False, figsize=(8,6))
```

```python
eda.box_plot(save=False)
```
