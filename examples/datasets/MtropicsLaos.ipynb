```python
laos = MtropicsLaos()
```

# Precipitation

```python
pcp = laos.fetch_pcp()
```

```python
pcp.shape
```

```python
eda = EDA(pcp, save=False)
```

```python
eda.plot_missing(figsize=(3,5))
```
