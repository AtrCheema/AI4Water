```python
import numpy as np

from ai4water.utils.datasets import arg_beach
from ai4water.experiments import MLRegressionExperiments
```

```python
# load the data
data = arg_beach()
data.head()
```

```python
# define input and output variables
inputs = list(data.columns)[0:-1]
outputs = [list(data.columns)[-1]]
```

```python
# we would like to transform the output on logarithmic scale
data['tetx_coppml'] = np.log10(data['tetx_coppml'].values)
```

```python
# All the arguments related to AI4Water's Model class will go here
comparisons = MLRegressionExperiments(data=data, input_features=inputs, output_features=outputs,
                                      transformation='minmax',
                                      val_fraction=0.0,
                                      val_data="same",
                                      test_fraction=0.3,
                                      verbosity=0,
                                      )
```

```python
# change the run_type to 'optimize' to optimize the hyperparameters of each model in experiment.
# For options related to hyperparameter optimization see help(comparisons.fit)
comparisons.fit(run_type="dry_run")
```

```python
# compare the R2 of different algorithms
m = comparisons.compare_errors('r2', save=True, fig_height=11)
m
```

```python
# compare the R2 of different algorithms
m = comparisons.compare_errors('mape', save=True, fig_height=11)
[val[1] for val in m.values()]
```

```python
# compare the R2 of different algorithms
m = comparisons.compare_errors('rmse', save=True, fig_height=11)
[val[1] for val in m.values()]
```

```python
# compare the R2 of different algorithms
m = comparisons.compare_errors('r2', save=True, fig_height=6, cutoff_val=0.5, cutoff_type='greater')
[val[1] for val in m.values()]
```

```python
comparisons.fit(run_type="optimize",
                opt_method="bayes",
                num_iterations=50,
                include=['XGBoostRegressor',
                         'NuSVR',
                         'RandomForestRegressor',
                         'XGBoostRFRegressor',
                         'ExtraTreesRegressor',
                         'HistGradientBoostingRegressor',
                         'SVR',
                         'CATBoostRegressor',
                         'LGBMRegressor',
                         'GradientBoostingRegressor',
                         'ADABoostRegressor',
                         'BaggingRegressor'
                ],
                post_optimize='train_best')
```

```python
comparisons.compare_errors('nse', save=True, name="optimized_nse", fig_height=5)
```

```python
comparisons.compare_errors('rmse', save=True, name="optimized_rmse", fig_height=5)
```

```python
comparisons.compare_errors('mape', save=True, name="optimized_mape", fig_height=5)
```

```python
comparisons.plot_taylor( 
             figsize=(7, 10),
             save=True,
             grid_kws={'axis': 'x', 'color': 'g', 'lw': 1.0},
            )
```
