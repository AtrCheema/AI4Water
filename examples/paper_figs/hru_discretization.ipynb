```python
import os

from AI4Water.utils.spatial_processing import MakeHRUs
```

```python
SLOPE = {0: '0-13',
         1: '13-26',
         2: '26-39',
         #3: '39-53'
         }

shapefile_paths = os.path.join(os.getcwd(), 'shapefiles')
Soil_shp = os.path.join(shapefile_paths, 'soil.shp')

SubBasin_shp = os.path.join(shapefile_paths, 'sub_basins.shp')
slope_shp = os.path.join(shapefile_paths, 'slope.shp')

years = {2011: os.path.join(shapefile_paths, 'lu2011.shp'),
         2012: os.path.join(shapefile_paths, 'lu2012.shp'),
         2013: os.path.join(shapefile_paths, 'lu2013.shp'),
         2014: os.path.join(shapefile_paths, 'lu2014.shp'),
         # 2015:"D:\\Laos\\data\\landuse\\shapefiles\\LU2015.shp"
         }
```

```python

hru_object = MakeHRUs('unique_lu_soil',
                      index=years,
                      subbasins_shape=SubBasin_shp,
                      soil_shape=Soil_shp,
                      slope_shape=slope_shp)

hru_object.call(plot_hrus=False)
```

```python
for yr in years:
    hru_object.draw_pie(yr, title=False, n_merge=4, save=True, textprops={'fontsize': '12'})
```

```python
hru_object.plot_as_ts(min_xticks=3, max_xticks=4, save=True)
```

```python
hru_object = MakeHRUs('unique_lu_soil_sub',
                      index=years,
                      subbasins_shape=SubBasin_shp,
                      soil_shape=Soil_shp,
                      slope_shape=slope_shp)

hru_object.call(plot_hrus=False)

for yr in years:
    hru_object.draw_pie(yr, title=False, n_merge=50, save=True, textprops={'fontsize': '12'})

hru_object.plot_as_ts(min_xticks=3, max_xticks=4, save=True)
```
